```python
from matplotlib import pyplot as plt
```

```python
import numpy
```

```python
vendas = numpy.array([10, 20, 30, 5, 8])
```

```python
periodo = numpy.array(["jan", "fev", "mar", "abr", "mai"])
```

```python
plt.plot(periodo, vendas)
plt.xlabel("Periodos")
plt.ylabel("Quatidade de vendas")
plt.title("Vendas no semestre")
plt.show()
```

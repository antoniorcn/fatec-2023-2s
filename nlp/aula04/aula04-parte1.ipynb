```python
import matplotlib.pyplot as plt
```

```python
valores_x = [1, 3, 5, 7, 9, 11, 13]
valores_y = [0, 2, 4, 6, 8, 10, 12]

plt.plot(valores_x, valores_y, marker="o", linestyle="-.")
```

```python
valores_y = [10, 20, 4, 6, 18, 100, 12]

plt.plot(valores_y, marker="o", linestyle="-.")
plt.ylabel("Vendas Mês")
plt.xlabel("Meses")
plt.text(4, 60, "Texto qualquer")
plt.title("Titulo do Grafico")
plt.suptitle("Titulo da figura")
# plt.savefig("imagem.png")
plt.legend()
```

```python
x = range(-100, 100, 10)

y1 = [elemento**2 for elemento in x]
y2 = [-elemento**2 for elemento in x]

plt.plot(x, y1, label="positivos")
plt.plot(x, y2, label="negativos")
plt.ylim(-2000, 2000)
plt.axhline(0)
plt.axvline(0)
plt.xlabel("Valores de X")
plt.ylabel("Valores de Y")
plt.legend()
```

```python
valores_x = [1.0, 2.0, 3.0]
valores_y = [2.0, 4.0, 1.0]
plt.plot(valores_x, valores_y, "-.*g")
plt.xlabel("eixo - x")
plt.ylabel("eixo - y")
plt.title("Meu primeiro gráfico")
```

```python
valores_x1 = [1.0, 2.0, 3.0]
valores_y1 = [2.0, 4.0, 1.0]
plt.plot(valores_x1, valores_y1, "-*g", label="vendas")

valores_x2 = [1.0, 2.0, 3.0]
valores_y2 = [4.0, 1.0, 3.0]
plt.plot(valores_x2, valores_y2, "-*r", label="compras")

plt.legend()
```

```python
plt.subplot(2, 2, 1)
valores_x1 = [1.0, 2.0, 3.0]
valores_y1 = [2.0, 4.0, 1.0]
plt.plot(valores_x1, valores_y1, "-*g", label="1Q")
plt.legend()
plt.title("Situação 1º quartil")

plt.subplot(2, 2, 2)
valores_x2 = [1.0, 2.0, 3.0]
valores_y2 = [4.0, 1.0, 3.0]
plt.plot(valores_x2, valores_y2, "-*r", label="2Q")
plt.legend()
plt.title("Situação 2º quartil")


plt.subplot(2, 2, 3)
valores_x1 = [1.0, 2.0, 3.0]
valores_y1 = [2.0, 4.0, 1.0]
plt.plot(valores_x1, valores_y1, "-*b", label="3Q")
plt.legend()
plt.title("Situação 3º quartil")

plt.subplot(2, 2, 4)
valores_x2 = [1.0, 2.0, 3.0]
valores_y2 = [4.0, 1.0, 3.0]
plt.plot(valores_x2, valores_y2, "-*y", label="4Q")
plt.legend()
plt.title("Situação 4º quartil")
plt.tight_layout()
```

```python
valores_x1 = ["jan", "fev", "mar"]
valores_y1 = [2.0, 4.0, 1.0]
plt.barh(valores_x1, valores_y1, label="vendas")

valores_x2 = ["jan", "fev", "mar"]
valores_y2 = [4.0, 1.0, 3.0]
plt.barh(valores_x2, valores_y2, label="compras")

plt.legend()
```

```python
import csv 

arquivo = open("./titanic.csv", "r", encoding="utf8")
arquivo_csv = csv.reader(arquivo, delimiter=",", quotechar='"')

idades = []

for indice, row in enumerate(arquivo_csv):
    if indice == 0:
        print(row)
    else:
        if row[5] != '':
            idade = int(float(row[5]))
            idades.append(idade)

print(idades)

bloco = 10
faixa = (20, 70)
           
plt.hist(idades, bloco, faixa, label="idades")
plt.legend()
```

```python
# ['Bom','dia','pessoal','vamos','nos','reunir','amanha']

def remover_pontuacao(texto):
    especiais = ['\n', '\r', '\t']
    texto_filtrado = ""
    for charactere in texto:
        if charactere not in punctuation:
            if charactere in especiais:
                texto_filtrado += ' '
            else:
                texto_filtrado += charactere
    return texto_filtrado
```

```python
def separar_tokens(texto):
    return texto.split(" ")
```

```python
def remover_stop_words(tokens):
    stopwords = nltk.corpus.stopwords.words('portuguese')
    token_sem_stop_words = []
    for token in tokens:
        nova_token = token.lower().strip()
        if nova_token != '' and nova_token not in stopwords:
            token_sem_stop_words.append(nova_token)
    return token_sem_stop_words
```

```python
arquivo = open("imdb-reviews-pt-br.csv", "r", encoding="utf8")
```

```python
arquivo_csv = csv.reader(arquivo, delimiter=",", quotechar='"')
```

```python
count = 0
sentencas = [
#    ['exemplo', 'motivo', 'maioria', 'filmes', 'ação'],
#    ['exemplo', 'motivo', 'maioria', 'filmes', 'ação'],
#    ['exemplo', 'motivo', 'maioria', 'filmes', 'ação']
]
todas_palavras = [
#    'exemplo', 'motivo', 'maioria', 'filmes', 
#    'ação', 'filme', 'costner', 'veio', 'assistir'
]

for row in arquivo_csv:
    
    texto_limpo = remover_pontuacao(row[2])
    tokens = separar_tokens(texto_limpo)
    tokens_sem_stop = remover_stop_words(tokens)
    
    todas_palavras.extend(tokens_sem_stop)
    
    sentencas.append(tokens_sem_stop)
    
    count += 1
    if count > 3:
        break
```

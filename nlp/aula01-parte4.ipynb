```python
from matplotlib import pyplot as plt
```

```python
import numpy
```

```python
from wordcloud import WordCloud
```

```python
texto = "A fatec é um lugar legal para estudar bastante"
```

```python
arr = texto.split(" ")
```

```python
arr
```

```python
vetor_palavras = numpy.array(arr)
```

```python
vetor_palavras
```

```python
# Nuvem de palavras
word_cloud_generator = WordCloud(width=800, height=600, max_font_size=32, collocations=False)
imagem1 = word_cloud_generator.generate(texto)
```

```python
plt.figure(figsize=(20, 15))
plt.imshow(imagem1, interpolation='bilinear')
plt.axis("off")
plt.show()
```
